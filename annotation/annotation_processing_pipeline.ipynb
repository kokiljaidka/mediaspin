```python
# Define a function to extract information from the 'prompt' column
def extract_info(prompt):
    lines = prompt.split('\n')
    original_headline = lines[0].replace('Original Headline: ', '')
    edited_headline = lines[1].replace('Edited Headline: ', '')
    added_words = lines[2].replace('Added words: ', '')
    removed_words = lines[3].replace('Removed words: ', '')
    return original_headline, edited_headline, added_words, removed_words

df[['original_headline', 'edited_headline', 'added_words', 'removed_words']] = df['prompt'].map(extract_info).apply(pd.Series)
```

```python
def column_stats_for_bias(df, col_name):
    none_count = df[col_name].str.count(r'\[None\]')
    added_count = df[col_name].str.count(r'\[Added\]')
    removed_count = df[col_name].str.count(r'\[Removed\]')
    print(f"'[None]': {none_count.sum()}")
    print(f"'[Added]': {added_count.sum()}")
    print(f"'[Removed]': {removed_count.sum()}")
    total_count = none_count.sum() + added_count.sum() + removed_count.sum()
    print(f"Sum: {total_count}")
    print(f"Length of df: {len(df)}")
```

```python
types_of_bias = ['spin', 'unsubstantiated_claims',
       'opinion_statements_presented_as_fact', 'sensationalism', 'mudslinging',
       'mind_reading', 'slant', 'flawed_logic', 'bias_by_omission',
       'omission_of_source_attribution', 'bias_by_story_choice_and_placement',
       'subjective_qualifying_adjectives', 'word_choice']
```

```python
for bias in types_of_bias:
    column_stats_for_bias(df, bias)
```
